```python
import pandas as pd
import matplotlib.pyplot as plt
import yfinance as yf
```

```python
tickers = ['LOW','TJX']

data = yf.download(tickers, period = '15y')['Close'][tickers]
```

```python
plt.figure(figsize=(12,6))
plt.plot(data[:50].index, data[:50]['LOW'], label='LOW')
plt.plot(data[:50].index, data[:50]['TJX'], label='TJX')
plt.legend()
plt.show()
```
